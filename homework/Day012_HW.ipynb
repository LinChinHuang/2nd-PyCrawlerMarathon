### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    #找出分類
    print(d.find(class_="tag").text, d.find_all('em')[-1].text)
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
dicSort = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    #找出分類
    sort = d.find_all('em')[-1].text
    lst = []
    if sort in dicSort :
        lst = dicSort[sort]
    dicSort[sort] = lst
    lst.append(d.find_all('a')[-1].text)

print(dicSort)
```

### ③ 哪一個類別的文章最多

```python
maxNewsCnt = 0
maxNewsSort = ""
for k,v in dicSort.items():
    if maxNewsCnt < len(v):
        maxNewsCnt = len(v)
        maxNewsSort = k
        
print("類別["+maxNewsSort+"]的文章最多，共"+str(maxNewsCnt)+"筆")
```
