```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = {}
currentKey = ""
for row in table.find_all('tr'):
    detectItem = row.find_all('td')
    
    if (len(detectItem) == 5): #含第一行(監測項目)
        if detectItem[0]:
            currentKey = detectItem[0].text
            childDic = {}
            childDic[detectItem[2].text] = detectItem[3].text
            d[currentKey] = childDic

    else:
        if (not currentKey == ""):
            #print(len(detectItem))
            d[currentKey][detectItem[0].text]=detectItem[1].text

print(d)
```
